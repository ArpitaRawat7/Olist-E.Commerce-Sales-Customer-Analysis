Products Table - Missing Value Investigation

A total of 610 products were found with missing category and descriptive attributes.

Investigation revealed that the same records were missing:
- product_category_name
- product_name_lenght
- product_description_lenght
- product_photos_qty

Since these products may still be referenced by orders and sales transactions, they were retained in the dataset rather than removed.

    Filling null records

```python
products['product_category_name'] = (
    products['product_category_name']
    .fillna('unknown')
)
```

```python
products['product_name_lenght'] = (
    products['product_name_lenght']
    .fillna(products['product_name_lenght'].median())
)

products['product_description_lenght'] = (
    products['product_description_lenght']
    .fillna(products['product_description_lenght'].median())
)

products['product_photos_qty'] = (
    products['product_photos_qty']
    .fillna(products['product_photos_qty'].median())
)
```

```python
for col in [
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]:
    products[col] = products[col].fillna(products[col].median())
```

    Removing Duplicates

We don't remove duplicates because multiple customers live in the same location.
We only remove rows that are completely identical across all columns.

```python
geo_before = geolocation.shape[0]

geolocation = geolocation.drop_duplicates()

geo_after = geolocation.shape[0]

print("Rows removed:", geo_before - geo_after)
```

The geolocation table contained exact duplicate records. These duplicates were removed to reduce redundancy while preserving all unique ZIP code-location mappings.

```python
customers['customer_city'] = (
    customers['customer_city']
    .str.lower()
    .str.strip()
)
```

```python
sellers['seller_city'] = (
    sellers['seller_city']
    .str.lower()
    .str.strip()
)
```

```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])
```
